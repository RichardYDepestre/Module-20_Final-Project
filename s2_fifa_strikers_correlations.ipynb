```python
# Initial imports.
%run 1st_Functions.ipynb

import numpy as np
from pathlib import Path
import os
```

```python
import os

strikers_dataset_loc= os.path.join('./resources', 'strikers.csv')
strikers_df = create_fifa_players_dataframe(strikers_dataset_loc)

# convert rating from numerical to categorical values
strikers_df['rating'] = (strikers_df['potential'].astype(int) + strikers_df['overall'].astype(int))/2
strikers_df['rating'] = strikers_df['rating'].apply(np.int64)

display(strikers_df)
```

# Define the features set

# Splitting into Train and Test sets

```python
#height Column
def clean_height(x):
    l=[]
    s=1
    if isinstance(x,str):
        l=x.split("'")
        i=int(l[0])
        j=int(l[1])
        s=((i*12)+j)/12
    return(s)


strikers_df['height'] = strikers_df['height'].apply(clean_height)
strikers_df['height'].fillna((strikers_df['height'].mean()), inplace = True)
```

```python
#weight Column
def clean_weight(x):
    
    if isinstance(x,str):
        return(x.replace('lbs', ''))
    return(x)

strikers_df['weight'] = strikers_df['weight'].apply(clean_weight).astype('float')
strikers_df['weight'].fillna((strikers_df['weight'].mean()), inplace = True)
```

```python
#strikers_df = encode_field(strikers_df) # 'field_position')

X = strikers_df.copy()
X = X.drop(["field_position", "height", "overall", "potential",'field_position'], axis=1)

Y = strikers_df['field_position'].values.ravel()

display(X)
display(Y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=78)
```

```python
from sklearn.preprocessing import StandardScaler

# Creating a StandardScaler instance.
scaler = StandardScaler()

# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# display(y_train)

display(X_train_scaled)
# display(X_test_scaled)
```

# Fit the Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier(n_estimators=32, random_state=42)

# Fitting the model
random_forest_model = random_forest_model.fit(X_train_scaled, y_train.ravel())
# display(random_forest_model)
```

# Making predictions using the Test Data

```python
predictions = random_forest_model.predict(X_test_scaled)
display(predictions)
```

# Evaluate the Model

```python
print(predictions)
```

```python
import pandas
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)
# display(cm)
# print(y_test.shape)
# print(y_test.shape)
y_test
predictions

# Create a DataFrame from the confusion matrix.
cm_df = pandas.DataFrame(cm, index=["Actual 1", "Actual 2", "Actual 3"], columns=["Predicted 1", "Predicted 2", "Predicted 3"])

display(cm_df)
```

# Calculating the accuracy score

```python
acc_score = accuracy_score(y_test, predictions)

# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
